```python
def get_price(url, headers, params):
  response = requests.get(url, params=params, headers=headers).text
  content = BeautifulSoup(response)
  price = int(content.find("h2").text[1:].replace(",", ""))
  sleep(1.5)
  return price
```

```python
prices = {}
for code in state_codes:
  url = f"https://www.zillow.com/{code}/home-values/"
  price = get_price(url, headers, params)
  prices[code] = price
```

```python
url = \
  "https://raw.githubusercontent.com/kjhealy/fips-codes/master/state_and_county_fips_master.csv"

fips_df = pd.read_csv(url)
fips_df["state"] = fips_df["state"].fillna(-1)
fips_df = fips_df[fips_df["state"] != -1]
fips_df["link"] = fips_df["name"] + " " + fips_df["state"]
fips_df["link"] = fips_df["link"].apply(lambda x: x.lower().replace(" ", "-"))
fips_df = fips_df[["name", "state", "link"]]
fips_df = fips_df.set_index(["state", "name"])
```

```python
state_codes_2 = state_codes[len(state_codes) // 2 : ]
```

```python
state_codes_2 = state_codes_2[18:]
```

```python
state_codes_2
```

```python

county_prices = {}

for code in state_codes_2:
  county_prices[code] = {}
  for cnty in fips_df.loc[code].index:
    url_code = fips_df.loc[code].loc[cnty]["link"]
    url = f"https://www.zillow.com/{url_code}/home-values/"
    price = get_price(url, headers, params)
    county_prices[code][cnty] = price
```
